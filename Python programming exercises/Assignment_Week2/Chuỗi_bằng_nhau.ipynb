**Python code**




```python
n=int(input())
count=0
for i in range(0,n):
	str1=str(input())
	str2=str(input())
	for j in range(0,len(str2)):
		if str2[j] in str1:
			count=count+1
	if count >0 and len(str1)==len(str2):
		print("YES")
	else:
		print("NO")
	count=0
```
