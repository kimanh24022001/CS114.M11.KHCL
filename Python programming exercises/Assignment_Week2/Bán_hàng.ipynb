```python
import math
q=int(input())
for i in range(0,q):
	sum=0
	n=int(input())
	for j in list(map(int, input().split())):
		sum=sum+j
	temp=int((sum/n))	
	while temp*n<sum:
		temp=temp+1
	if temp*n>=sum:
		temp=temp
	print(temp)    	# your code goes here
```
